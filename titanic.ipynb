```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head(50)
```

```python
X = titanic.drop('PassengerId', axis=1).values
y = titanic['PassengerId'].values
```

```python
X_Age = X[:,4]
type(X_Age), type(y)
```

```python
y = y.reshape(-1,1)
X_Age = X_Age.reshape(-1,1)
```

```python
plt.scatter(X_Age, y)
plt.ylabel('Passenger Id')
plt.xlabel('Age')
plt.show()
```

```python
model = LinearRegression
```

```python
model.fit(X_Age, y)
```

```python
prediction_space = np.linspace(min(X_Age), max(X_Age)).reshape(-1,1)
```

```python
plt.scatter(X_Age, y, color='green')
plt.plot(prediction_space, model.predict(prediction_space),
        color='red', linewidth=3)
plt.show()
```
